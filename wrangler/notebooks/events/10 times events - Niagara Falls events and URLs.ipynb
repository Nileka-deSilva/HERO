# 1) Scrape all event names and event URLs

```python
from bs4 import BeautifulSoup # Import for Beautiful Soup
import requests # Import for requests
import lxml # Import for lxml parser
import datetime

main_link = "https://10times.com/niagarafalls-ca"


headers = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36'}

# Request for the URL
page = requests.get(main_link, headers=headers)

# Make it a soup
soup = BeautifulSoup(page.text,'html.parser')


event_name_list = soup.find(class_='col-md-5')


# Pull text from all instances of <a> tag within BodyText div
event_name_list_items = event_name_list.find_all(['a','div'])
```

```python
for event_name in event_name_list_items:
    names = event_name.get('data-ga-label')
    links = event_name.get('href')
    startDate= event_name.get('data-start-date')
    endDate=event_name.get('data-end-date')
    
    print(names)
    print(links)
    print(startDate)
    print(endDate)  
```

```python
import csv
from csv import writer

with open('10timesEventLinks.csv', 'a', encoding='utf8', newline='') as f:
    fieldnames = ['Event name', 'URL']
    thewriter = csv.DictWriter(f, fieldnames=fieldnames)
    
    
    for event_name in event_name_list_items:
        names = event_name.get('data-ga-label')
        links = event_name.get('href')
        
        
        thewriter.writeheader()
        thewriter.writerow({'Event name': names, 'URL': links})
    
           
```

# 2) Create and clean data frame with event names and URLs - into CSV file 
 

```python
import pandas as pd
import numpy as np
df = pd.read_csv('10timesEventLinks.csv')
df
```

```python
df = df.dropna()
df= df[df["Event name"].str.contains("Event name")==False]
df["Event name"] = df["Event name"].apply(lambda x: x.replace("To", ""))
df = df.reset_index(drop=True)
df
```

```python
df=df.drop_duplicates()
df = df.reset_index(drop=True)
```
