## Sentence Embeddings

* [Finding semantic similarities between sentences](https://enjoymachinelearning.com/blog/finding-semantic-similarity-between-sentences-in-python/)
* [SBERT Semantic Textual Similarity](https://www.sbert.net/docs/usage/semantic_textual_similarity.html)

```python
emb_kwargs = {
    "LOWER":True,
    "NO_STOP_WORDS":False,
    "METRIC":'COSIN',
    "MAX_SCORES":2,
    "TOLERANCE":0.7,
}
sim_scores_df = clsNLP.get_similarity_scores(
    input_sentences = list(data.room_type.sample(n=5)),
    lookup_sentences =list(_room_taxo.description),
    **emb_kwargs)
sim_scores_df.score=sim_scores_df.score.astype('float64')
sim_scores_df=sim_scores_df.reset_index()
sim_scores_df = sim_scores_df[~(sim_scores_df['score'] <= emb_kwargs['TOLERANCE'])]
```

```python
_selected_df = pd.DataFrame()
if sim_scores_df.shape[0] > 0:
    sim_scores_df['room_cate'] = sim_scores_df.groupby('input_sentence')['score'].\
                            transform('idxmax').map(sim_scores_df['lookup sentence'])
    for rtype in sim_scores_df.input_sentence.unique():
        best_guess = sim_scores_df.loc[sim_scores_df.input_sentence == rtype]
        best_guess = best_guess.loc[best_guess.score.idxmax()]
        _guess_dict = {'room_type':best_guess.input_sentence,
                       'room_cate':best_guess.room_cate,
                       'score':best_guess.score,
                      }
#         print(best_guess.input_sentence)
        _selected_df = pd.concat([_selected_df,pd.DataFrame([_guess_dict])])
else:
    print("no data")
_selected_df.to_csv(os.path.join(ROOT_DIR,'data/tmp/similarity_scores.csv'))
```

```python
import nltk

def remove_adj(sentence):
    
    stop_tags = ["JJ", "JJR", "JJS", "NN"]
    tokens = nltk.word_tokenize(sentence)
    tags = list(reversed(nltk.pos_tag(tokens)))
    noun_located = False
    stop_reached = False
    final_sent = ''

    for word,pos in tags:
        if noun_located == False and pos == 'NN':
            noun_located = True
            final_sent+=f' {word}'
        elif stop_reached == False and pos in stop_tags:
            stop_reached = True
        elif stop_reached == True:
            final_sent+=f' {word}'

    final_sent = ' '.join(reversed(final_sent.split(' ')))      
    return final_sent

x = remove_adj('what is the mean of profit')
print(x)
```
